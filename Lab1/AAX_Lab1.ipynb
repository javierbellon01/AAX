The logs of the nodes contain the relative power they have to each other, so we can parse them to obtain the power they detect between them.

```python
# Parsing
import re

logs_path = "/content/drive/Shareddrives/AAX/Lab1/output"

def read_prx_from_node(filename):

  with open(filename, 'r') as f:
    file_content = f.read()
    file_lines = file_content.split("\n")

    Prx_node = {}

  for i in range(len(file_lines)):
    if re.search("(N\d+ has started a TX of packet #\d+ \(type \d+\) to N\d+)", file_lines[i]):
      node_line = file_lines[i].split(";")[4]
      power_line = file_lines[i+2].split(";")[4]

      node = re.search("N\d+", node_line).group(0)
      power = re.search("-?\d+.?\d*", power_line).group(0)

      Prx_node[node] = float(power)

  if (len(Prx_node) == 0):
    print("No transmissions found, check if the input file is correct")
    return False

  return Prx_node

Prx_AP_A = read_prx_from_node(f"{logs_path}/logs_output__N0_AP_A.txt")
print(f"N0: {Prx_AP_A}")
Prx_STA_A1 = read_prx_from_node(f"{logs_path}/logs_output__N1_STA_A1.txt")
print(f"N1: {Prx_STA_A1}")
Prx_AP_B = read_prx_from_node(f"{logs_path}/logs_output__N2_AP_B.txt")
print(f"N2: {Prx_AP_B}")
Prx_STA_B1 = read_prx_from_node(f"{logs_path}/logs_output__N3_STA_B1.txt")
print(f"N3: {Prx_STA_B1}")
```

We then plot the representation of the network in a 2D space, with the power relation between them.

```python
import matplotlib.pyplot as plt
import numpy as np

# Node positions (meters)
positions = {
    'AP_A': (0, 0),                         # AP_A at 0, 0
    'STA_A1': (0, 4),  # STA_A1 at 0, 0 + distance
    'AP_B': (5, 0),                         # AP_B at 5, 0
    'STA_B1': (5, 4)   # STA_A1 at 5, 0 + distance
}

# Channel used per node (input your values here)
channels = {
    'AP_A': 0,
    'STA_A1': 0,
    'AP_B': 0,
    'STA_B1': 0
}

# Perceived power strength by each STA from its AP
#perceived_power_dbm = {
 #   ('AP_A', 'STA_A1'): Prx_STA_A1['N0'],   # N0 = AP_A
  #  ('AP_B', 'STA_B1'): Prx_STA_B1['N2']    # N2 = AP_B
#}

perceived_power_dbm = {
  ('AP_A', 'STA_A1'): Prx_AP_A['N1'],   # N0 = AP_A.  N1 = STA_A1
  ('AP_B', 'STA_B1'): Prx_AP_B['N3'],   # N2 = AP_B.  N3 = STA_B1
  ('AP_A', 'STA_B1'): Prx_AP_A['N3'],
  ('STA_A1', 'AP_B'): Prx_AP_B['N1'],
  ('AP_A', 'AP_B'): Prx_AP_A['N2'],
  ('STA_A1', 'STA_B1'): Prx_STA_A1['N3']
}

# Plot setup
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_title("Intra-BSS Wireless Scenario with Channels and Perceived Power")
ax.set_xlabel("X (m)")
ax.set_ylabel("Y (m)")
ax.set_aspect('equal')

# Draw nodes
for node, (x, y) in positions.items():
    color = 'blue' if 'AP' in node else 'red'
    ax.plot(x, y, 'o', color=color, markersize=10)
    ax.text(x + 0.1, y + 0.3, f"{node}\nCh {channels[node]}", fontsize=9)

# Intra-BSS links only
links = [('AP_A', 'STA_A1'), ('AP_B', 'STA_B1'), ('AP_A', 'STA_B1'), ('AP_A', 'AP_B'), ('STA_A1', 'STA_B1'), ('STA_A1', 'AP_B')]

for tx, rx in links:
    x1, y1 = positions[tx]
    x2, y2 = positions[rx]

    # Draw arrow
    ax.annotate("",
                xy=(x2, y2), xytext=(x1, y1),
                arrowprops=dict(arrowstyle="<->", color='green', lw=2))

    # Label perceived power
    power = perceived_power_dbm.get((tx, rx), None)
    if power is not None:
        mid_x, mid_y = (x1 + x2) / 2, (y1 + y2) / 2
        ax.text(mid_x + 0.2, mid_y, f"{power} dBm", fontsize=9, color='darkgreen')

# Plot appearance
ax.grid(True)
plt.xlim(-1, 7)
plt.ylim(-1, 6)
plt.tight_layout()
plt.show()
```

The different nodes form a rectangle, where we can confirm the power between two nodes is relative to the distance between them.

Our algorithm takes elements from 𝛆-greedy and Thompson sampling. The algorithm implements 𝛆 with decay to incentivize exploration in early stages and decrease the probability of exploration the further the algorithm iterates.

On exploration: We create a custom estimate for each arm based on its average reward but punish according to a ratio between the current iteration number and times that arm has been explored, so less explored arms will be less punished. We take the action corresponding to the best estimate.

On exploitation: We create a custom estimate for each arm based on the sampling from a manufactured normal distribution; as in Thompson sampling, the standard deviation is computed exactly the same, but we take a single sample and not k samples summed. We take the action corresponding to the best estimate.

On both cases, we check for available arms, and if there's no available arm after finishing a count, we just pick a random action, in both cases, to avoid deterministically choosing the best action, aiming to prevent collisions.

As expected, the result is similar to ɛ-greedy. This time, however, our initial exploration phase consciously takes what could be suboptimal choices instead of pure random. Eventually the options are well explored, and the known reward probably outweighs the penalization inserted.

Our algorithm seems to stably reach wider periods of higher throughput earlier by having a brief period of high oscillation between high and low throughput due to our deterministic part (custom exploration estimate exploring the least explored actions). However, we can see a lot more oscillations during all the simulation, probably due to our availability check leading to random choices in both exploration and exploitation if no actions are available.

It has the same range as ɛ-greedy, although it appears from looking at the plots that ɛ-greedy is more stable overall, even if it converges later.

Next, we plot the throughput means and max-min values per AP.

```python
# (response to exercise 5)
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data_egreedy_1 = {"AP 0" : list(A0_throughput_egreedy_1.values()),
                  "AP 1": list(A1_throughput_egreedy_1.values()),
                  "AP 2": list(A2_throughput_egreedy_1.values())}

df_throughput_egreedy_1 = pd.DataFrame(data_egreedy_1)

data_ml4net_1 = {"AP 0" : list(A0_throughput_ml4net_1.values()),
                 "AP 1": list(A1_throughput_ml4net_1.values()),
                 "AP 2": list(A2_throughput_ml4net_1.values())}

df_throughput_ml4net_1 = pd.DataFrame(data_ml4net_1)

labels = ("ɛ-greedy", "ML4Net")

combined_means = {
    "AP 0": (df_throughput_egreedy_1['AP 0'].mean(), df_throughput_ml4net_1['AP 0'].mean()),
    "AP 1": (df_throughput_egreedy_1['AP 1'].mean(), df_throughput_ml4net_1['AP 1'].mean()),
    "AP 2": (df_throughput_egreedy_1['AP 2'].mean(), df_throughput_ml4net_1['AP 2'].mean())
}

combined_maxmin = {
    "AP 0": (df_throughput_egreedy_1['AP 0'].max() - df_throughput_egreedy_1['AP 0'].min(), df_throughput_ml4net_1['AP 0'].max() - df_throughput_ml4net_1['AP 0'].min()),
    "AP 1": (df_throughput_egreedy_1['AP 1'].max() - df_throughput_egreedy_1['AP 1'].min(), df_throughput_ml4net_1['AP 1'].max() - df_throughput_ml4net_1['AP 1'].min()),
    "AP 2": (df_throughput_egreedy_1['AP 2'].max() - df_throughput_egreedy_1['AP 2'].min(), df_throughput_ml4net_1['AP 2'].max() - df_throughput_ml4net_1['AP 2'].min())
}
```

```python
x = np.arange(len(labels))
width = 0.25

fig, ax = plt.subplots(layout='constrained')
fig.set_figheight(6)
fig.set_figwidth(8)

for i, (label, throughput_data) in enumerate(combined_means.items()):
    rects = ax.bar(x + i * width, throughput_data, width, label=label)
    ax.bar_label(rects, padding=3)

ax.set_ylabel('Throughput (Mbps)')
ax.set_title("Average AP throughput for different simulations")
ax.set_xticks(x + width, labels)
ax.legend()

plt.show()
```

As expected from the throughput plots, our algorithm has a lower average throughput compared to ε-greedy for all APs. This difference is greater than when we had two APs, so it seems our algorithm is less reliable the greater the AP number.

```python
x = np.arange(len(labels))
width = 0.25

fig, ax = plt.subplots(layout='constrained')
fig.set_figheight(6)
fig.set_figwidth(8)

for i, (label, throughput_data) in enumerate(combined_maxmin.items()):
    rects = ax.bar(x + i * width + width / 2, throughput_data, width, label=label)
    ax.bar_label(rects, padding=3)

ax.set_ylabel('Throughput (Mbps)')
ax.set_title("Difference between maximum and minimum throughput for different simulations")
ax.set_xticks(x + width, labels)
ax.legend()

plt.show()
```

Our custom algorithm scores a lower difference between the maximum and minimum points for all APs, except for 2, in which it is the same. Considering the throughput plots, the only AP that experiences a collision (throughput is 0) is AP 2, so it makes sense for it to be the highest max-min difference in both simulations. Because of this observation, we can infer that ε-greedy has the ability to reach a higher maximum throughput.

For AP 0 and 2, the difference is lower in ML4Net. It could either be because the decreases in throughput of ε-greedy reach lower levels, or because ε-greedy tends to perform slightly better.


We have seen in detail how both algorithms perform with 3 APs, so next, we will run all the seeds and comment on the general results.

```python
seeds = (3, 5, 11, 13)

throughputs_egreedy = []
throughputs_ml4net = []

for seed in seeds:
  print(f"Starting executions for seed {seed}\n")
  print(f"E-greedy execution")

  # Run e-greedy simulation again
  INPUT_FILE_NODES='../input/input_nodes_3.csv'
  INPUT_FILE_AGENTS='../input/agents_egreedy_3.csv'
  OUTPUT_FILE_LOGS='../output/logs_test_agents_egreedy_3.txt'
  SEED=seed

  !cd $mypath/main; ./komondor_main $INPUT_FILE_NODES $INPUT_FILE_AGENTS $OUTPUT_FILE_LOGS $SIMULATION_CODE $FLAG_SAVE_NODE_LOGS $FLAG_SAVE_AGENT_LOGS $FLAG_PRINT_SYSTEM_LOGS $FLAG_PRINT_NODE_LOGS $FLAG_PRINT_AGENT_LOGS $SIM_TIME $SEED

  egreedy_object = {
      "seed": seed,
      "AP 0": get_throughput_agent(f"{logs_path}/logs_output_AAX-LAB1-AGENTS_A0_A.txt"),
      "AP 1": get_throughput_agent(f"{logs_path}/logs_output_AAX-LAB1-AGENTS_A1_B.txt"),
      "AP 2": get_throughput_agent(f"{logs_path}/logs_output_AAX-LAB1-AGENTS_A2_C.txt")
      }

  throughputs_egreedy.append(egreedy_object)

  # Run ml4net simulation again
  print(f"ml4net execution")
  INPUT_FILE_AGENTS='../input/agents_m4net_3.csv'
  OUTPUT_FILE_LOGS='../output/logs_test_agents_m4net_3.txt'
  SEED=seed

  !cd $mypath/main; ./komondor_main $INPUT_FILE_NODES $INPUT_FILE_AGENTS $OUTPUT_FILE_LOGS $SIMULATION_CODE $FLAG_SAVE_NODE_LOGS $FLAG_SAVE_AGENT_LOGS $FLAG_PRINT_SYSTEM_LOGS $FLAG_PRINT_NODE_LOGS $FLAG_PRINT_AGENT_LOGS $SIM_TIME $SEED

  ml4net_object = {
      "seed": seed,
      "AP 0": get_throughput_agent(f"{logs_path}/logs_output_AAX-LAB1-AGENTS_A0_A.txt"),
      "AP 1": get_throughput_agent(f"{logs_path}/logs_output_AAX-LAB1-AGENTS_A1_B.txt"),
      "AP 2": get_throughput_agent(f"{logs_path}/logs_output_AAX-LAB1-AGENTS_A2_C.txt")
      }

  throughputs_ml4net.append(ml4net_object)
```

Next, we can plot the results.

```python
for i in range(len(seeds)):

  seed = seeds[i]
  egreedy_obj = throughputs_egreedy[i]
  ml4net_obj = throughputs_ml4net[i]

  # More visual separation of each seed
  print("##################################################################################")
  print(f"######################## STARTING EXECUTIONS WITH SEED {seed} #########################")
  print("##################################################################################")
  print("\n")

  # Plot throughput (e-greedy)
  print(f"Throughput plot for e_greedy, seed={seed}")
  plotting_throughput([egreedy_object["AP 0"], egreedy_object["AP 1"], egreedy_object["AP 2"]])
  print("\n")

  # Plot throughput (ml4net)
  print(f"Throughput plot for ml4net, seed={seed}")
  plotting_throughput([ml4net_object["AP 0"], ml4net_object["AP 1"], ml4net_object["AP 2"]])
  print("\n")

  # organize data
  data_egreedy_N = {"AP 0": list(egreedy_obj["AP 0"].values()),
                    "AP 1": list(egreedy_obj["AP 1"].values()),
                    "AP 2": list(egreedy_obj["AP 2"].values())}

  df_throughput_egreedy_N = pd.DataFrame(data_egreedy_N)

  data_ml4net_N = {"AP 0": list(ml4net_obj["AP 0"].values()),
                  "AP 1": list(ml4net_obj["AP 1"].values()),
                  "AP 2": list(ml4net_obj["AP 2"].values())}

  df_throughput_ml4net_N = pd.DataFrame(data_ml4net_N)

  labels = ("ɛ-greedy", "ML4Net")

  combined_means = {
      "AP 0": (df_throughput_egreedy_N['AP 0'].mean(), df_throughput_ml4net_N['AP 0'].mean()),
      "AP 1": (df_throughput_egreedy_N['AP 1'].mean(), df_throughput_ml4net_N['AP 1'].mean()),
      "AP 2": (df_throughput_egreedy_N['AP 2'].mean(), df_throughput_ml4net_N['AP 2'].mean())
  }

  combined_maxmin = {
      "AP 0": (df_throughput_egreedy_N['AP 0'].max() - df_throughput_egreedy_N['AP 0'].min(), df_throughput_ml4net_N['AP 0'].max() - df_throughput_ml4net_N['AP 0'].min()),
      "AP 1": (df_throughput_egreedy_N['AP 1'].max() - df_throughput_egreedy_N['AP 1'].min(), df_throughput_ml4net_N['AP 1'].max() - df_throughput_ml4net_N['AP 1'].min()),
      "AP 2": (df_throughput_egreedy_N['AP 2'].max() - df_throughput_egreedy_N['AP 2'].min(), df_throughput_ml4net_N['AP 2'].max() - df_throughput_ml4net_N['AP 2'].min())
  }

  print("Mean and minmax throughput comparison plots")
  # Plot mean data
  x = np.arange(len(labels))
  width = 0.25

  fig, ax = plt.subplots(layout='constrained')
  fig.set_figheight(6)
  fig.set_figwidth(8)

  for i, (label, throughput_data) in enumerate(combined_means.items()):
      rects = ax.bar(x + i * width, throughput_data, width, label=label)
      ax.bar_label(rects, padding=3)

  ax.set_ylabel('Throughput (Mbps)')
  ax.set_title(f"Average AP throughput for different simulations, seed={seed}")
  ax.set_xticks(x + width, labels)
  ax.legend()

  plt.show()

  x = np.arange(len(labels))
  width = 0.25

  fig, ax = plt.subplots(layout='constrained')
  fig.set_figheight(6)
  fig.set_figwidth(8)

  # Plot max-min data
  for i, (label, throughput_data) in enumerate(combined_maxmin.items()):
      rects = ax.bar(x + i * width + width / 2, throughput_data, width, label=label)
      ax.bar_label(rects, padding=3)

  ax.set_ylabel('Throughput (Mbps)')
  ax.set_title(f"Difference between maximum and minimum throughput for different simulations, seed={seed}")
  ax.set_xticks(x + width, labels)
  ax.legend()

  plt.show()

  print("\n")
```

There are a few interesting things happening.

Contrary to what we have seen before, with seed 3, the max-min difference is bigger for ML4Net in AP 0 and 1 and the same for AP 2. That means that with our custom algorithm we can actually reach the same maximum throughput as with ε-greedy because, once again, AP 2 reaches a minimum point of 0.

AP 0 and 1 being lower than ε-greedy also means that there may be executions where our algorithm presents the same or better stability, even if it does not have the higher average throughput or it has a difficult beginning phase. This is supported by the observations of the diagrams of all the seeds, but it could also be a byproduct of randomness.
